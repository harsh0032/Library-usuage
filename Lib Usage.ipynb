```python
pivot = pd.pivot_table(df,
            values = ['Total Checkouts'],
            index = ['Patron Type Definition'], 
                       columns= ['Age Range'],
                       aggfunc=[np.mean], 
                       margins=True)
pivot
cmap = sns.cubehelix_palette(start = 1.5, rot = 1.5, as_cmap = True)
plt.subplots(figsize = (30, 20))
sns.heatmap(pivot,linewidths=0.2,square=True )

#Majorits of entries in the category of the patron type 
#defination and the age group have total checkouts of 250 and below

#retired staff has total checkouts of more than 1750

```

```python
pivot = pd.pivot_table(df,
            values = ['Total Renewals'],
            index = ['Patron Type Definition'], 
                       columns= ['Age Range'],
                       aggfunc=[np.mean], 
                       margins=True)
pivot
cmap = sns.cubehelix_palette(start = 1.5, rot = 1.5, as_cmap = True)
plt.subplots(figsize = (30, 20))
sns.heatmap(pivot,linewidths=0.2,square=True )

#Reitred staff of age 20-24 has more than 2000 Renewals 

```

```python
pivot = pd.pivot_table(df,
            values = ['Total Renewals'],
            index = ['Patron Type Definition'], 
                       columns= ['Circulation Active Month'], 
                       margins=True)
pivot
cmap = sns.cubehelix_palette(start = 1.5, rot = 1.5, as_cmap = True)
plt.subplots(figsize = (30, 20))
sns.heatmap(pivot,linewidths=0.2,square=True )

#The highest value of the total renewals was recevied in the month of july in the patron type of 'Staff'


```
